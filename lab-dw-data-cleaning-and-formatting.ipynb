## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:
- *Column names should be in lower case*
- *White spaces in column names should be replaced by `_`*
- *`st` could be replaced for `state`*

```python
# Your code here
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
data.head()

```

```python
# 1 - Column names should be in lower case
data.columns = data.columns.str.lower()
data
```

```python
# 2 - *White spaces in column names should be replaced by `_`
data.columns = data.columns.str.replace(' ', '_')
data
```

```python
# 3 -`st` could be replaced for `state`

data.rename(columns={'st': 'state'}, inplace=True)
data
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python

# 1 Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*

data["gender"].unique() # first we nead to see how many different values there are
```

```python

# exemple 1 - data['gender'].replace({'F': 'F', 'Femal': 'F', 'female': 'F', 'M': 'M', 'Male': 'M'}, inplace=True)

# Instead of using "exemple 1" that is limited to what exists, let's try to generalize a littel more for any word starting with f we will change to f and any word starting with m to change to m
```

```python

data['gender'].fillna('unknown', inplace=True)  # After getting some errors we came to the conclusion that it would be necessary to fill in the NaN fields


data['gender'] = data['gender'].str.lower()  # In order to standardize, we change all words to lowercase letters.



# And then we replace all words that start with 'f' to 'F'  all words that start with 'm' to 'M'

data.loc[data['gender'].str.startswith('f'), 'gender'] = 'F'
data.loc[data['gender'].str.startswith('m'), 'gender'] = 'M'


data["gender"].unique()
```

```python
# 2 - State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
data["state"].unique()
```

```python
data['state'].replace({'Cali': 'California', 'AZ': 'Arizona', 'WA': 'Washington'}, inplace=True)
data["state"].unique()
```

```python
# 3 - In education, "Bachelors" could be replaced by "Bachelor"*

data["education"].unique()
```

```python
data['education'].replace({'Bachelors': 'Bachelor'}, inplace=True)
data["education"].unique()
```

```python
# 4 - In Customer Lifetime Value, delete the `%` character*
data["customer_lifetime_value"]
```

```python
data["customer_lifetime_value"] = data["customer_lifetime_value"].str.rstrip('%')
data["customer_lifetime_value"]
```

```python
# 5 - In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*
data["vehicle_class"].unique()
```

```python
data["vehicle_class"].replace({"Sports Car": "Luxury","Luxury SUV": "Luxury","Luxury Car": "Luxury"}, inplace=True)
data["vehicle_class"].unique()
```
